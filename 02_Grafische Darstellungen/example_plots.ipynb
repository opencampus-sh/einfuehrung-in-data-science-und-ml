```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Load the mpg dataset from seaborn
mpg = sns.load_dataset('mpg')

# First, let's see what columns we actually have
print("Available columns in the dataset:")
print(mpg.columns.tolist())
```

```python
# Histogram of MPG distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=mpg, x='mpg', bins=20)
plt.title('Distribution of Miles Per Gallon')
plt.xlabel('Miles Per Gallon (MPG)')
plt.ylabel('Count')
plt.show()
```

```python
# Bar plot of average MPG by number of cylinders
plt.figure(figsize=(12, 6))
sns.barplot(data=mpg, x='cylinders', y='mpg')
plt.title('Average MPG by Number of Cylinders')
plt.xlabel('Number of Cylinders')
plt.ylabel('Average MPG')
plt.show()
```

```python
# Scatter plot of MPG vs displacement, colored by cylinders
plt.figure(figsize=(10, 6))
sns.scatterplot(data=mpg, x='displacement', y='mpg', hue='cylinders', palette='viridis')
plt.title('MPG vs Engine Displacement by Number of Cylinders')
plt.xlabel('Engine Displacement (cubic inches)')
plt.ylabel('Miles Per Gallon')
plt.show()
```

```python
# Faceted scatter plots by origin
plt.figure(figsize=(15, 5))
g = sns.FacetGrid(mpg, col='origin', col_wrap=3)
g.map(sns.scatterplot, 'displacement', 'mpg')
g.fig.suptitle('MPG vs Displacement by Origin', y=1.05)
plt.show()
```
