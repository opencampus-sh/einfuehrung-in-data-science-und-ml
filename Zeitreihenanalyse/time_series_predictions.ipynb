### Prepare Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# Reading the data file
pedestrians_hourly = pd.read_csv("https://raw.githubusercontent.com/opencampus-sh/einfuehrung-in-data-science-und-ml/main/Zeitreihenanalyse/kiel-holstenstra%C3%9Fe%20(mitte)-20200505-20220103-day.csv", sep=';')
pedestrians_daily = pd.read_csv("https://raw.githubusercontent.com/opencampus-sh/einfuehrung-in-data-science-und-ml/main/Zeitreihenanalyse/kiel-holstenstra%C3%9Fe%20(mitte)-20200505-20220103-hour.csv", sep=';')

# Prepare data
pedestrians_hourly['datetime'] = pd.to_datetime(pedestrians_hourly['time of measurement'])
pedestrians_daily['date'] = pd.to_datetime(pedestrians_daily['time of measurement'])
```

### Pedestrians hourly

```python
# Basic plot
plt.figure(figsize=(10, 6))
plt.plot(pedestrians_hourly['datetime'], pedestrians_hourly['pedestrians count'], color="#69b3a2")
plt.xticks(rotation=60, ha='right')
plt.xlabel("")
plt.tight_layout()
plt.show()
```

```python
# Time frame specific plot
plt.figure(figsize=(10, 6))
plt.plot(pedestrians_hourly['datetime'], pedestrians_hourly['pedestrians count'], color="#69b3a2")
plt.xticks(rotation=60, ha='right')
plt.xlabel("")
plt.xlim(pd.to_datetime("2021-10-01"), pd.to_datetime("2021-11-01"))
plt.tight_layout()
plt.show()
```

### Pedestrians daily

```python
# Basic plot
plt.figure(figsize=(10, 6))
plt.plot(pedestrians_daily['date'], pedestrians_daily['pedestrians count'], color="#69b3a2")
plt.xticks(rotation=60, ha='right')
plt.xlabel("")
plt.tight_layout()
plt.show()
```

```python
# Time frame specific plot
plt.figure(figsize=(10, 6))
plt.plot(pedestrians_daily['date'], pedestrians_daily['pedestrians count'], color="#69b3a2")
plt.xticks(rotation=60, ha='right')
plt.xlabel("")
plt.xlim(pd.to_datetime("2021-10-01"), pd.to_datetime("2021-11-01"))
plt.tight_layout()
plt.show()
```

```python
# Another time frame specific plot
plt.figure(figsize=(10, 6))
plt.plot(pedestrians_daily['date'], pedestrians_daily['pedestrians count'], color="#69b3a2")
plt.xticks(rotation=60, ha='right')
plt.xlabel("")
plt.xlim(pd.to_datetime("2021-01-01"), pd.to_datetime("2021-12-31"))
plt.tight_layout()
plt.show()
```

### Time series including differencing

```python
# Create the differenced time series (weekly differencing)
differenced_weekly = pedestrians_daily['pedestrians count'].diff(periods=7).dropna()
```

```python
# Differenced time series plot
plt.figure(figsize=(10, 6))
plt.plot(pedestrians_daily['date'][7:], differenced_weekly, color="#69b3a2")
plt.xticks(rotation=60, ha='right')
plt.xlabel("")
plt.xlim(pd.to_datetime("2021-01-01"), pd.to_datetime("2021-12-31"))
plt.tight_layout()
plt.show()
```

```python
# Including constrained y-scale
plt.figure(figsize=(10, 6))
plt.plot(pedestrians_daily['date'][7:], differenced_weekly, color="#69b3a2")
plt.xticks(rotation=60, ha='right')
plt.xlabel("")
plt.xlim(pd.to_datetime("2021-01-01"), pd.to_datetime("2021-12-31"))
plt.ylim(-10000, 10000)
plt.tight_layout()
plt.show()
```
