```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the CSV file into a DataFrame
df = pd.read_csv("umsatzdaten_gekuerzt.csv")

# Group the data by 'Warengruppe' and calculate the sum of 'Umsatz' for each group
grouped_data = df.groupby('Warengruppe')['Umsatz'].sum()

# Calculate mean and standard deviation for each group
mean_values = df.groupby('Warengruppe')['Umsatz'].mean()
std_values = df.groupby('Warengruppe')['Umsatz'].std()

# Calculate confidence intervals (95% confidence level)
confidence_intervals = 1.96 * (std_values / np.sqrt(df.groupby('Warengruppe')['Umsatz'].count()))

# Plot the bar chart with confidence intervals
grouped_data.plot(kind='bar', yerr=confidence_intervals, capsize=5)

# Add labels and title
plt.xlabel('Warengruppe')
plt.ylabel('Total Umsatz')
plt.title('Total Umsatz by Warengruppe with 95% Confidence Intervals')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 1. Den Datensatz "umsatzdaten_gekuerzt.csv" einlesen
df = pd.read_csv("umsatzdaten_gekuerzt.csv")

# Um sicherzustellen, dass das Datum als Datum interpretiert wird
df['Datum'] = pd.to_datetime(df['Datum'])

# 2. Den Durchschnittsumsatz je Wochentag berechnen
df['Wochentag'] = df['Datum'].dt.day_name()  # Wochentag aus Datum extrahieren
mean_sales_by_weekday = df.groupby('Wochentag')['Umsatz'].mean()

# 3. Konfidenzintervalle der Umsätze je Wochentag berechnen (95% Konfidenzniveau)
std_sales_by_weekday = df.groupby('Wochentag')['Umsatz'].std()
confidence_intervals = 1.96 * (std_sales_by_weekday / np.sqrt(df.groupby('Wochentag')['Umsatz'].count()))

# 4. Die Wochentage von Montag bis Sonntag ordnen
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
mean_sales_by_weekday = mean_sales_by_weekday.reindex(weekday_order)
confidence_intervals = confidence_intervals.reindex(weekday_order)

# 5. Balkendiagramm mit Durchschnittsumsätzen je Wochentag und Konfidenzintervallen erstellen
plt.bar(mean_sales_by_weekday.index, mean_sales_by_weekday, yerr=confidence_intervals, capsize=5)

# Achsenbeschriftungen und Titel hinzufügen
plt.xlabel('Wochentag')
plt.ylabel('Durchschnittlicher Umsatz')
plt.title('Durchschnittlicher Umsatz je Wochentag mit Konfidenzintervallen (95% CI)')

# Diagramm anzeigen
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 1. Read the dataset "umsatzdaten_gekuerzt.csv"
df = pd.read_csv("umsatzdaten_gekuerzt.csv")

# To ensure the date is interpreted as a date
df['Datum'] = pd.to_datetime(df['Datum'])

# 2. Calculate the average sales per weekday
df['Weekday'] = df['Datum'].dt.day_name()  # Extract weekday from date
mean_sales_by_weekday = df.groupby('Weekday')['Umsatz'].mean()

# 3. Calculate confidence intervals of sales per weekday (95% confidence level)
std_sales_by_weekday = df.groupby('Weekday')['Umsatz'].std()
confidence_intervals = 1.96 * (std_sales_by_weekday / np.sqrt(df.groupby('Weekday')['Umsatz'].count()))

# 4. Order the weekdays from Monday to Sunday
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
mean_sales_by_weekday = mean_sales_by_weekday.reindex(weekday_order)
confidence_intervals = confidence_intervals.reindex(weekday_order)

# 5. Create bar chart with average sales per weekday and confidence intervals
plt.bar(mean_sales_by_weekday.index, mean_sales_by_weekday, yerr=confidence_intervals, capsize=5)

# Add axis labels and title
plt.xlabel('Weekday')
plt.ylabel('Average Sales')
plt.title('Average Sales per Weekday with Confidence Intervals (95% CI)')

# Show the plot
plt.show()
```
